# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class Kumanda:
    def __init__(self, tv_durum="Kapalı", tv_ses=10, kanal_listesi=["TRT"], kanal="TRT"):
        self.tv_durum = tv_durum
        self.tv_ses = tv_ses
        self.kanal_listesi = kanal_listesi
        self.kanal = kanal

    def tv_ac(self):
        if self.tv_durum == "Açık":
            print("Televizyon zaten açık.")
        else:
            print("Televizyon açılıyor...")
            time.sleep(1)
            self.tv_durum = "Açık"
            print("Televizyon açıldı.")

    def tv_kapat(self):
        if self.tv_durum == "Kapalı":
            print("Televizyon zaten kapalı.")
        else:
            print("Televizyon kapatılıyor...")
            time.sleep(1)
            self.tv_durum = "Kapalı"
            print("Televizyon kapandı.")

    def ses_ayarları(self):
        while True:
            secim = input("Sesi Azalt: '-' Sesi Artır: '+' Çıkış: 'q'\nSeçiminiz: ")
            if secim == "-":
                if self.tv_ses > 0:
                    self.tv_ses -= 1
                    print(f"Ses: {self.tv_ses}")
                else:
                    print("Ses zaten en düşük seviyede.")
            elif secim == "+":
                if self.tv_ses < 100:
                    self.tv_ses += 1
                    print(f"Ses: {self.tv_ses}")
                else:
                    print("Ses zaten en yüksek seviyede.")
            elif secim == "q":
                print("Ses ayarından çıkılıyor...")
                break
            else:
                print("Geçersiz işlem!")

    def kanal_ekle(self):
        yeni_kanal = input("Eklemek istediğiniz kanalın adı: ")
        print("Kanal ekleniyor...")
        time.sleep(1)
        self.kanal_listesi.append(yeni_kanal)
        print(f"{yeni_kanal} kanalı eklendi.")

    def kanal_sayisi(self):
        print(f"Toplam kanal sayısı: {len(self.kanal_listesi)}")

    def rastgele_kanal(self):
        self.kanal = random.choice(self.kanal_listesi)
        print(f"Şu anki kanal: {self.kanal}")

    def televizyon_bilgileri(self):
        print(f"TV Durumu: {self.tv_durum}")
        print(f"TV Ses: {self.tv_ses}")
        print(f"Kanal Listesi: {self.kanal_listesi}")
        print(f"Şu Anki Kanal: {self.kanal}")

kumanda = Kumanda()

print("""
Televizyon Kumandası Programı
-----------------------------
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğren
6. Rastgele Kanala Geç
7. Televizyon Bilgileri
Programdan çıkmak için 'q' tuşlayınız.
""")

while True:
    islem = input("Yapmak istediğiniz işlem: ")

    if islem == "q":
        print("Program sonlandırılıyor...")
        break
    elif islem == "1":
        kumanda.tv_ac()
    elif islem == "2":
        kumanda.tv_kapat()
    elif islem == "3":
        kumanda.ses_ayarları()
    elif islem == "4":
        kumanda.kanal_ekle()
    elif islem == "5":
        kumanda.kanal_sayisi()
    elif islem == "6":
        kumanda.rastgele_kanal()
    elif islem == "7":
        kumanda.televizyon_bilgileri()
    else:
        print("Geçersiz işlem! Lütfen tekrar deneyin.")
```
